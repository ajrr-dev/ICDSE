```python
# Different types of values and operators in Python

# Integer value: A whole number without a decimal point
x = 10
# Float value: A number that has a decimal point
y = 3.14
# String value: A sequence of characters enclosed in quotes (either single or double quotes)
name = "Alice"
# Boolean value: A data type with only two possible values: True or False
is_student = True

# Arithmetic operators: Symbols that perform mathematical operations

# Addition: Adds two numbers
sum_result = x + y  # 10 + 3.14 = 13.14
# Subtraction: Subtracts the second number from the first
difference = x - y  # 10 - 3.14 = 6.86
# Multiplication: Multiplies two numbers
product = x * y  # 10 * 3.14 = 31.4
# Division: Divides the first number by the second
quotient = x / y  # 10 / 3.14 ≈ 3.18

# Print the results: Display the output of the arithmetic operations
print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)

# Comparison operators: Symbols that compare two values and return a Boolean result (True or False)

# Greater than: Checks if the value on the left is greater than the value on the right
print("Is x greater than y?", x > y)  # 10 > 3.14 → True
# Equal to: Checks if the value on the left is equal to the value on the right
print("Is x equal to 10?", x == 10)  # 10 == 10 → True

# Logical operators: Symbols that combine multiple Boolean expressions and return a Boolean result

# Logical AND: Returns True if both conditions are True
print("Is x greater than 5 and y less than 4?", x > 5 and y < 4)  # 10 > 5 and 3.14 < 4 → True and True → True


# Integer value: A whole number without a decimal point
x = 1

# String value: A sequence of characters enclosed in quotes (either single or double quotes)
y = "1"

# Arithmetic operation with integer
sum_result = x + x  # 1 + 1 = 2 (Adding two integer values)

# Casting: The process of converting a value from one data type to another
# Here, str(x) converts the integer value x to a string so it can be concatenated with another string
concat_result = str(x) + y  # "1" + "1" = "11" (Combining two string values)

# Print the results: Display the output of the arithmetic and concatenation operations
print("Sum of 1 + 1 as integers:", sum_result)  # Output: 2
print('Concatenation of "1" + "1" as strings:', concat_result)  # Output: 11

# Demonstrating type difference

# type() function: Returns the data type of the variable
print("Type of x:", type(x))  # <class 'int'> (x is an integer)
print("Type of y:", type(y))  # <class 'str'> (y is a string)
```
